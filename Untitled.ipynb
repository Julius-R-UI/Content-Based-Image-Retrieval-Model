```python
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
```

```python
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(512,512,3))
base_model.trainable = False
```

```python
input_img = layers.Input((512,512,3))
l = layers.Conv2D(8, (3,3), activation = tf.keras.layers.LeakyReLU(0.25), padding = 'same')(input_img)
l = layers.MaxPooling2D((2,2))(l)
l = layers.Conv2D(16, (3,3), activation = tf.keras.layers.LeakyReLU(0.25), padding = 'same')(l)
l = layers.MaxPooling2D((2,2))(l)
l = layers.Conv2D(32, (3,3), activation = tf.keras.layers.LeakyReLU(0.25), padding = 'same')(l)
l = layers.MaxPooling2D((2,2))(l)
l = layers.Conv2D(64, (3,3), activation = tf.keras.layers.LeakyReLU(0.25), padding = 'same')(l)
l = layers.MaxPooling2D((2,2))(l)
l = layers.Conv2D(128, (3,3), activation = tf.keras.layers.LeakyReLU(0.25), padding = 'same')(l)
l = layers.Flatten()(l)
encoded = layers.Dense(128, activation = tf.keras.layers.LeakyReLU(0.25))(l)

l = layers.Dense(32*32*128, activation = tf.keras.layers.LeakyReLU(0.25))(encoded)
l = layers.Reshape((32,32,128))(l)
l = layers.Conv2D(128, (3, 3), activation=tf.keras.layers.LeakyReLU(0.25), padding='same')(l)
l = layers.UpSampling2D((2, 2))(l)
l = layers.Conv2D(64, (3, 3), activation=tf.keras.layers.LeakyReLU(0.25), padding='same')(l)
l = layers.UpSampling2D((2, 2))(l)
l = layers.Conv2D(32, (3, 3), activation=tf.keras.layers.LeakyReLU(0.25), padding='same')(l)
l = layers.UpSampling2D((2, 2))(l)
l = layers.Conv2D(16, (3, 3), activation=tf.keras.layers.LeakyReLU(0.25), padding='same')(l)
l = layers.UpSampling2D((2, 2))(l)
l = layers.Conv2D(8, (3, 3), activation=tf.keras.layers.LeakyReLU(0.25), padding='same')(l)
decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same', kernel_initializer = tf.keras.initializers.GlorotNormal())(l)

#b = base_model(decoded)
auto_encoder = Model(input_img, decoded)
```

```python
def display(x, chosen_locations):
    plt.figure(figsize=(10,10)) # specifying the overall grid size

    for i in range(len(chosen_locations[0])):
        plt.subplot(5,5,i+1)    # the number of images in the grid is 5*5 (25)
        plt.imshow(x[chosen_locations[0][i]])
        
    plt.show()
```

```python
def knn(layer,n, image, db):
    encoder_ = Model(auto_encoder.input,auto_encoder.layers[layer].output)
    pred = encoder_.predict(x)
    
    knn = NearestNeighbors(n_neighbors=n, metric="cosine")
    knn.fit(pred)
    _, ranks = knn.kneighbors(encoder_.predict(image.reshape((1,512,512,3))))
    return ranks
```

```python
for i in x_t:
    ranks = knn(11, 5, i, x)
    plt.imshow(i)
    display(x,ranks)
```
